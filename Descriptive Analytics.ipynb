###  Grouping and Aggregating

#### What is the average SOLD PRICE for each age category?

```python
ipl_auction_df.groupby('AGE')['SOLD PRICE'].mean()
```

```python
soldprice_by_age=ipl_auction_df.groupby('AGE')['SOLD PRICE'].mean().reset_index()
print(soldprice_by_age)
```

#### Average SOLD PRICE for Dierent Playing Roles in Each Age Category?

```python
soldprice_by_age_role=ipl_auction_df.groupby(['AGE','PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()
print(soldprice_by_age_role)
```

### Joining DataFrames

#### Compare the average auction price for dierent ages and playing roles.


```python
soldprice_comparision=soldprice_by_age_role.merge(soldprice_by_age,how="outer",on="AGE")
```

```python
soldprice_comparision
```

#### Renaming Columns

```python
soldprice_comparision.rename(columns={'SOLD PRICE_x':'SOLD_PRICE_AGE_ROLE','SOLD PRICE_y':'SOLD_PRICE_AGE'},inplace=True)
```

```python
soldprice_comparision.head()
```

### Applying Operations to multiple column

#### Percentage change in SOLD PRICE

```python
soldprice_comparision['CHANGE']=soldprice_comparision.apply(lambda rec: (rec.SOLD_PRICE_AGE_ROLE-rec.SOLD_PRICE_AGE)/rec.SOLD_PRICE_AGE,axis=1)
```

```python
print(soldprice_comparision)
```

### Box Plot

```python
box = sns.boxplot(ipl_auction_df['SOLD PRICE']);
```

```python
# To obtain min,max 1st quartile (25 percentile), 3rd quartile (75 percentile) in the boxplot

box = plt.boxplot(ipl_auction_df['SOLD PRICE']); 
```

```python
[item.get_ydata()[0] for item in box['caps']]  # caps in the box variable returns the min and the max value
```

```python
[item.get_ydata()[0] for item in box['whiskers']]  # whiskers in the box variable returns the 1st quartile (25 percentile) and 3rd quartile (75 percentile) value
```

```python
# IQR= Q3-Q1 => 700000-225000=475000
```

```python
[item.get_ydata()[0] for item in box['medians']] 
```
